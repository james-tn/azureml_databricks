```
#Read data from Azure Open Source Data SOurce for year 2008

ids_weather = spark.read.format("parquet").load("abfss://mltraining@adlsdatalakegen6.dfs.core.windows.net/ISDWeather/year=2008/")
```

```
%sql  SET spark.databricks.delta.symlinkFormatManifest.fileSystemCheck.enabled = false
```

```
spark.conf.set(
  "fs.azure.account.key.adlsdatalakegen6.dfs.core.windows.net",
  dbutils.secrets.get(scope="scope1",key="adlsdatalakegen6"))
```

```
ids_weather.write.format("delta").partitionBy("year", "month").save("abfss://mltraining@adlsdatalakegen6.dfs.core.windows.net/ISDWeatherDelta")
```

```
%sql ALTER TABLE delta.`abfss://mltraining@adlsdatalakegen6.dfs.core.windows.net/ISDWeatherDelta` SET TBLPROPERTIES(delta.compatibility.symlinkFormatManifest.enabled=true)
```
